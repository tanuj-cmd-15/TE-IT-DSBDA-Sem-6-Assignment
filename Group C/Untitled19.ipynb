```python
rating=var2.find('div',{'class':'_2d4LTz'})
rating.text
```

```python
def indrating(var2):
    data_string=" "
    data_list=[]
    for irw in var2.find_all('div',{'class':'_3LWZlK _1BLPMq'}):
        data_string=irw.get_text()
        data_list.append(data_string)
        data_string=" "
    return data_list
irating=indrating(var2)
print(irating)
```

```python
def cust_name(var2):
    data_string=" "
    data_list=[]
    for irw in var2.find_all('p',{'class':'_2sc7ZR _2V5EHH'}):
        data_string=irw.get_text()
        data_list.append(data_string)
        data_string=" "
    return data_list
cust_data=cust_name(var2)
print(cust_data)
```

```python
def review(var2):
    data_string=" "
    data_list=[]
    for irw in var2.find_all('div',{'class':""}):
        data_string=irw.get_text()
        data_list.append(data_string)
        data_string=" "
    return data_list
allreview=review(var2)
print(allreview)
```
