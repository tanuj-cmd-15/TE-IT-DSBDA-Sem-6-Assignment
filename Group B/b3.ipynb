#### Replacing the null values

```python
A["Ozone"]=A["Ozone"].fillna(A["Ozone"].mean())
A["Solar.R"]=A["Solar.R"].fillna(A["Solar.R"].mean())
```

```python
A
```

```python
A.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
A
```

#### PIE CHART

```python
A.columns
```

```python
labels = ['Ozone','Solar.R','Wind','Temp']
```

```python
Sizes= [A['Ozone'].mean(),A['Solar.R'].mean(),A['Wind'].mean(),A['Temp'].mean()]
```

```python
colors=['red','green','yellow','orange']
```

```python
explode=(0.1,0,0,0)
```

```python
plt.pie(Sizes,explode=explode,labeldistance=1.1,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('Average Data')
plt.show()
```

#### Bar Plot

```python
def addlabels():
    for i in range(len(A.iloc[0:21,3])):
        plt.text(i,A.iloc[0:21,3][i],A.iloc[0:21,3][i],ha='center')
y=np.arange(len(A.iloc[0:21,3]))
plt.figure(figsize=(10,6))
addlabels()
plt.bar(x=y,height=A.iloc[0:21,3],tick_label=y,color="dodgerblue")
plt.title('Wind measure for 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Wind')
plt.show()
```

#### BOX Plot

```python
A.boxplot(column=['Ozone','Wind','Temp','Solar.R'],grid=True)
plt.title('Box-Plot')
plt.suptitle('Temperature')
plt.ylabel('range')
plt.savefig('plot3.png')
```

#### HISTROGRAM

```python
plt.figure(figsize=(10,6))
plt.hist(x=A['Temp'],bins=20,color="Red",label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("count")
plt.plot()
```

#### SCATTER Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(A.iloc[0:30,5],A.iloc[0:30,2],color="green")
plt.title("Wind measures of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

#### PAIR Plot

```python
import seaborn 
seaborn.pairplot(A.iloc[:,[0,1,2,3,6]],hue='Humidity')
plt.show()
```

#### HEATMAP

```python
seaborn.heatmap(data=A.iloc[0:30,[0,1,2,3]])
plt.show()
```

#### Line graph

```python
plt.figure(figsize=(10,4))
plt.plot(A.iloc[0:30,5],A.iloc[0:30,3],color="blue")
plt.title('Temperature of May month')
plt.xlabel("Day")
plt.ylabel('Temperature')
```
