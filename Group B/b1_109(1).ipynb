# Create a data subset


```python
sub1=A[['Type','Category']]
sub1
```

```python
sub2 = A.loc[0:4]
sub2
```

```python
sub3= A.iloc[[0,1,3,6],[0,1,2]]
sub3
```

# Merge Dataset

```python
m1=sub2.merge(sub3)
m1.head(5)
```

```python
m3 = sub1.merge(sub2,left_index=True,right_index=True)
m3
```

```python
m3 = sub2.merge(sub3,on = None)
m3.head(5)
```

# Sort Data

```python
st1 =A.sort_values(by ='Type')
st1.head(5)
```

```python
st2 =A.sort_values(by =['Type', 'Category'])
st2.head(5)
```
